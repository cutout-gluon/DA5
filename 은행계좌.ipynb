```python
import random as r

class Account():      # 01.Account 클래스를 생성
    은행이름 = 'SC은행'  # 01. 은행 브랜드 이름 지정
    count = 0         # Q2. 계좌 객체의 개수를 저장

    def __init__(self): # 생성자 구현 (계좌개설): 입력받는 용도가 아닌 외부에서 값을 받아서 저장하는 방식
        self.예금주 = input("예금주 이름을 입력하세요: ")
        self.잔액 = int(input("예금액을 입력하세요: __원"))
        
        num1 = r.randint(100,999)
        num2 = r.randint(10,99)
        num3 = r.randint(100000,999999)
        self.계좌번호 = f"{num1}-{num2}-{num3}" 
        self.입금횟수 = 0   # Q7. 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가
        self.total_history = []             # Q10. 입급과 출금 내역이 기록, (거래종류, 거래금액, 잔액) 튜플 저장
        Account.count += 1          

    def __str__(self):
        return f"------------\n은행: {Account.은행이름}\n예금주: {self.예금주}\n계좌번호: {self.계좌번호}\n잔액: {self.잔액:,}"                  

    @classmethod
    def get_account_num(cls):   
        print(f"현재까지 개설된 총 계좌의 수는 {cls.count}개입니다.")  # Q3. 계좌의 개수를 출력



    def deposit(self, amount):                          # Q4. 입금 메서드

        if amount > 0:
            self.잔액 += amount
            self.입금횟수 += 1        # Q7. 입금 횟수 5회차 1%의 이자가 잔고에 추가
            self.total_history.append(("입금", amount, self.잔액))
            print(f"{self.예금주}님의 계좌에 {amount}원이 입금되었습니다.") #현재 잔액은{self.잔액}원입니다.

        if self.입금횟수 == 5:          
            print(f"5회 입금 보너스!: 이자 {self.잔액*0.01:,.0f}원이 입금되었습니다.")
            self.잔액 *= 1.01
            self.total_history.append(("이자", int(self.잔액 * 0.01),int(self.잔액 * 1.01)))
            self.입금횟수 = 0  # 횟수 리셋
            print(f"현재 잔액은 {self.잔액:,.0f}원 입니다.")

    def withdraw(self, amount):
        if amount <= self.잔액:
            self.잔액 -= amount
            self.total_history.append(("출금", amount, self.잔액))
            print(f"{amount}을 출금하였습니다. 잔액은 {self.잔액:,}입니다.")
        elif amount > self.잔액:
            print(f"잔액이 부족합니다.")
    
    def displ__info(self):  #정보출력 메서드
        print(f"은행:{Account.은행이름}, 예금주:{self.예금주}, 계좌번호:{self.계좌번호}, 잔액:{self.잔액:,}\n")





            
```

```python
계좌_목록 = []

계좌1 = Account()                # Q3. 계좌 개설
계좌_목록.append(계좌1)

계좌2 = Account()                # Q8. Account 클래스로부터 3개 이상 인스턴스를 생성
계좌_목록.append(계좌2)

계좌3 = Account()
계좌_목록.append(계좌3)
```

```python
print(계좌1)
print(계좌2)
print(계좌3)
```

```python
Account.displ__info(계좌1)                  # Q6. 정보 출력 메서드
Account.displ__info(계좌2)
Account.displ__info(계좌3)
```
