```python
import random as r

# Q1. Character 클래스 (모든 캐릭터의 기본 설계도)
class Chracter:
    def __init__(self, 이름, 레벨, 체력, 공격력, 방어력):
        self.이름 = 이름
        self.레벨 = 레벨
        self.체력 = 체력
        self.공격력 = 공격력
        self.방어력 = 방어력

    # 살아있는지 확인 (체력이 0보다 크면 True 반환)
    def is_alive(self):
        return self.체력 > 0
 
    # 데미지를 입는 로직 (받은 데미지 - 방어력)
    def take_damage(self, damage):
        actual_damage = max(0, damage - self.방어력) # 데미지가 방어력보다 작아도 최소 0 유지
        self.체력 -= actual_damage
        print(f"{self.이름}이(가) {actual_damage}만큼 데미지를 입었습니다! (남은 체력: {max(0, self.체력)})")
   
    # 타겟(대상)을 공격하는 로직
    def attack_target(self, target):
        damage = r.randint(1, self.공격력) # 1부터 내 공격력 사이의 랜덤 데미지
        print(f"\n{self.이름}이(가) {target.이름}을 공격합니다!")
        target.take_damage(damage) # 상대방의 take_damage 메서드 호출

# Q2-1. Player 클래스 (Character 상속)
class Player(Chracter):
    def __init__(self, 이름):
        # super()를 사용하여 부모 클래스의 생성자에 기본 수치 전달
        super().__init__(이름, 1, 100, 25, 5)    
        self.경험치 = 0 # 플레이어만 가지는 고유 속성
     
    def gain_experience(self, amount):
        self.경험치 += amount
        print(f"경험치가 {amount}만큼 증가하였습니다. (현재 경험치: {self.경험치})")

    def level_up(self):
        # 경험치가 50 이상일 때 능력치 상승
        if self.경험치 >= 50:
            self.레벨 += 1
            self.공격력 += 10
            self.방어력 += 5
            self.경험치 -= 50
            print(f"★ 레벨업! 현재 레벨: {self.레벨} ★")

# Q2-2. Monster 클래스 (Character 상속)
class Monster(Chracter):
    def __init__(self, 이름, 레벨):
        # 생성될 때 레벨에 맞춰 랜덤한 능력치를 계산
        체력 = r.randint(10, 30) * 레벨
        공격력 = r.randint(5, 20) * 레벨
        방어력 = r.randint(1, 5) * 레벨
        # 부모 생성자에게 이름, 레벨 및 계산된 랜덤 능력치 전달
        super().__init__(이름, 레벨, 체력, 공격력, 방어력)

# Q3. battle 함수 (독립 함수 - 캐릭터와 몬스터의 중재자)
def battle(player, monster):
    print(f"\n--- {monster.이름}(Lv.{monster.레벨})이(가) 나타났습니다! ---")
    
    # 두 인스턴스가 모두 살아있는 동안 반복
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster) # 플레이어 선공
        
        if not monster.is_alive(): # 몬스터가 죽으면 즉시 중단
            break
            
        monster.attack_target(player) # 몬스터 반격

    # 결과 처리
    if player.is_alive():
        print(f"\n[전투 승리!] {monster.이름}을(를) 물리쳤습니다.")
        player.gain_experience(monster.레벨 * 20)
        player.level_up()
        return True # 승리 시 True 반환
    else: 
        print(f"\n[전투 패배..] {player.이름}이(가) 쓰러졌습니다.")
        return False # 패배 시 False 반환

# Q4. main 함수 (게임 전체 흐름 제어)
def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    
    # 1. 플레이어 생성 (입력받은 이름으로 실제 객체 제작)
    p_name = input("플레이어의 이름을 입력하세요: ")
    player = Player(p_name) 
    
    # 2. 몬스터 딕셔너리를 순회하며 전투 진행
    for m_name, m_level in monster_dict.items():
        # 매 반복마다 새로운 몬스터 인스턴스 생성
        monster = Monster(m_name, m_level)
        
        # 전투 시작 및 결과 저장
        is_victory = battle(player, monster)
        
        # 3. 만약 전투에서 졌다면 게임 즉시 종료
        if not is_victory:
            print("\n--- 게임오버 ---")
            return 
            
    print("\n축하합니다! 모든 시련을 이겨내고 전설이 되었습니다.")

# 프로그램 실제 실행 지점 (왼쪽 벽에 붙어있어야 함)
if __name__ == "__main__":
    main()
```
